# Retrieve Rows by Index Position with .iloc[]
.iloc() is short for index location and if we're dealing with a typical numeric index. 
This function works as the same with loc() but here the index is numerical type.
iloc()-->It extracts based on index position so it accepts integers as its arguments.
loc()-->It extracts based on index position so it accepts STRINGS as its arguments.


```python
bond=pd.read_csv("jamesbond.csv")
bond.head(3)
```

```python
bond.loc[15]
bond.iloc[15]

bond.iloc[[15,20]] # displays the both the indexes.
bond.iloc[:5]  # displays the 0-4 indexes.
bond.iloc[15:] #displays 15- last.
bond.iloc[4:8+1] #displays 4-7 ie 4 values.
```

# The Catch-All .ix[] Method
 catch-all is because the I-X method basically combines the functionalities of the Loc and ILO methods.
 So it effectively captures all of the possible scenarios whether we want to extract rows from a data
frame based on the index label or the next position.
It basically allows you to do either.
So it's kind of this optimal method for when you kind of want to mix and match index labels or index

```python
bond=pd.read_csv("jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
# functions as .loc()
bond.ix["GoldenEye"]
bond.ix[["Diamonds are Forever","Moonraker","Spectre"]]
bond.ix["A View to a Kill":] 
# functions as .iloc()
bond.ix[0]
bond.ix[[1,5,4]]
bond.ix[5:9] 
bond.ix[5:19] 
bond.ix[5:] 
```

# random sample with .sample method()
  
    This method is used to make samples of columns or rows.
  Tablename.sample(n,frac,axis) where n is number of rows and frac is percnetage of data frame extraction and axis shows 0,1

```python
bond=pd.read_csv("jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

# FILTERING WITH THE .WHERE() METHOD

    Here we are getting the filtering method to extract at some condition.
    by the function .where()
    inside the where we can execute many of functions with and,or keyword functions.

```python
bond=pd.read_csv("jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
# this returns only specific rows which obeys condition
mask1=bond["Actor"]=="Sean Connery"  
mask2=bond["Budget"]>80
bond[mask1]   
bond[mask2]
```

```python
# this returns all data frame where  specific rows which obeys condition are values and all are NAN vlaues
bond.where(mask1)
bond.where(mask2)
# executing the both conditions by and,or key words as mask1 and mask2
bond.where(mask1 & mask2)
```

# THE .query() method

    here the qrguments taken are strings and doest work out if it has the space between column names.
    inside the QUERY() we write the condition and to separate the internal and external string we use "" and '' for that.
    inside query we can write the where type condition, this query method  works more effectently in large dataframes.
    tablename.query(" condition --> column/index name="key"')
    Here query ---> aa question.

```python
bond=pd.read_csv("jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
bond.columns=[column_name.replace(" ","_") for column_name in bond.columns]
# the above statement replace the space with the underscore in column names.
bond.head(1)
```

```python
bond.query("Actor=='Sean Connery'") 
# returns the only rows as condition actor is Sean connery
bond.query('index=="Casino Royale"')
# returns the index as the conditional statement.
```

```python
# this returns the rows based on condition.
bond.query("Director == 'Terence Young'")
bond.query("Actor !='Roger Moore'")
```

```python
# In this we can even pass the list for query condition
bond.query("Actor !=['Roger Moore','Terence Young']")
bond.query("Budget >100 and  Actor!=['Roger Moore','Terence Young']")
bond.query("Budget >100 or  Actor == ['Roger Moore','Terence Young']")
# Complex operator querys may also be done here.
bond.query("Budget<100 and Box_Office>350 or Year<1999 and  Actor!=['Roger Moore','Terence Young']")
```

# THE .apply() Methods in columns

    Here,the apply method is used for the functions in the data frames.
    Here, we are passing the single columns into the function
    Takes the column as series beacause same data type.

```python
bond=pd.read_csv("jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
def convert_to_string_add_millions(number):
    return str(number)+ "MILLIONS!"
bond["Box Office"]=bond["Box Office"].apply(convert_to_string_add_millions)
bond
```

```python
# ie passing the columns to function and mauplating the rows by function, return type is data frame.
bond["Budget"]=bond["Budget"].apply(convert_to_string_add_millions)
bond
```

```python
columns=["Box Office","Budget","Bond Actor Salary"]
for col in columns:
    bond[col]=bond[col].apply(convert_to_string_add_millions)
```

```python
bond.head(3)
```

# THE .apply() method in rows

    Here,the apply method is used for the functions in the data frames.
    Here, we are passing the rows into the function
    Takes the row elements as list beacause different data types.

```python
bond=pd.read_csv("jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
# ie passing the one by one rows to function and mauplating the columns by fucntion , return type is list
def good_movie(row):
    actor=row[1]
    budget=row[4]
    
    if budget>100.5:
        return "High Budget film"
    elif budget<60.5 and actor==["Roger Moore","Terence Young"]:
        return "High  Budget film"
    elif budget<60.5 and actor==["Roger Moore","Terence Young"]:
        return "Low budget Film"
    else:
        return "Medium Budget Film"

bond.apply(good_movie,axis="columns") 
```

# THE .copy() method
   
    This method is used to copy the existing data frame (or) series with exactly sepearte memoryy.every operation we done in copy donot change the data in original data frame.

```python
bond=pd.read_csv("jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
# Traditional and standard way of copying data ,which when manuplated changes data in original data frame.
directors=bond["Director"]
directors.head(3)
```

```python
directors.head(3) # in directors the value is effected by before statement
```

```python
# To prevent the above condition we use .copy method instead of directly assining the values to variables
# whatever operations we perform in copy donot effect the original array.

```
