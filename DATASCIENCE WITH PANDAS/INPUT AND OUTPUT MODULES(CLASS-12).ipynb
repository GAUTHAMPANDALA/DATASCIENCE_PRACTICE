# Export DataFrame to CSV File with the .to_csv() Method

# EXPORT THE EXCEL FILES

```python
url="https://data.cityofnewyork.us/api/views/2r9r-m6j4/rows.csv" 
chs=pd.read_csv(url)
chs.head()
```

```python
year_11=chs[chs["Year"]==2011]
year_14=chs[chs["Year"]==2014]
```

```python
excel_file=pd.ExcelWriter("years.xlsx") #creating the excel file
```

```python
#putting the sheets into file.
year_11.to_excel(excel_file, sheet_name="YEAR-11",index=False,columns=["Upper95%CI","Prevalence","Survey"])
year_14.to_excel(excel_file, sheet_name="YEAR-14",index=False, columns=["Lower95%CI","Prevalence","Survey"])
```

```python
excel_file.save() #saving the excel file containing the sheets.
```
