# CATEGORY 
rather than use a string to store a unique string and every single cell it points every cell that has the exact same value to a single object.IE, ALL THE  UNIQUE VALUES ARE CORRESPONDED TOOTHER COLUMNS WITH THEIR RESPECTIVE VALUE.This reduces the size(a->1,2,5,6,7... and b->599,55,44,..) like that

```python
nba["Position"]=nba["Position"].astype("category") 
```

```python
nba["Team"]=nba["Team"].astype("category")
```

```python
nba.dtypes
nba.info()  
# after converting float -int, categoring the duplicate values we have decreased memory
# meomory is decreased from 35.1kb to 25.8kb
```

# RANK VALUES WITH THE .RANK() METHOD
 Generates the brand new series of ranks. This method will execute if there are no NAN values , ie all NAN values should be converted to numbers

```python
nba=pd.read_csv("nba.csv").dropna(how="all")
nba["Salary"]=nba["Salary"].fillna(0).astype("int")
nba.head()
```

```python
nba["Salary Rank"]=nba["Salary"].rank(ascending=False).astype("int")
nba.head(5)
# gives the rank of salary ie, rank 361 for index 1 like that and HIGH SALLRY GET LOW RANK
```

```python
nba.sort_values(by="Salary",ascending=False,inplace=True) # sorting according to salary rank.
nba
```
