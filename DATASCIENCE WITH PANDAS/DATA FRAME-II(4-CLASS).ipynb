# BOTH AND & OR  COMBINED
      We have to make sure about ("paranthesie") for respective condition to make pandas understand input.
   IN pandas The given argument check rows one by one, if any one of condition "or" Respective "and" condition is satisfied then the output returns the row based on that.

```python
mask1=df["Team"]=="Client Services"
mask2=df["Start Date"]>"2006-01-01"
mask3= df["Senior Management"]
mask4=df["First Name"]=="Robert"
df[(mask3 & mask1) |mask2 & mask4 ]
# First the inner brackets is executed and after that the other two masks is obtained
# the below output returns rows which has :
#"TEAM AS MARKETING " OR "START DATE <1982-01-01" AND BOTH "Senior managment, Larry name"
```

# The .duplicatemethod()
This method makes the first occured value as not duplicate and all other same values occured after first is set as duplicate.
This consists of keep atribute 1.first 2.last 3.False

```python
df=pd.read_csv("employees.csv",parse_dates=["Start Date","Last Login Time"])
df["Senior Management"]=df["Senior Management"].astype(bool)
df["Gender"]=df["Gender"].astype("category")
df.sort_values("First Name",inplace=True)
df.head(3)
```

```python
df[df["First Name"].duplicated()]
```

```python
df["First Name"].duplicated(keep="last")
# keeps the first occured values at last
# where keep=first values keep first occured values at first.
```

```python
df[df["First Name"].duplicated(keep=False)]
# this marks all repeated as Duplicate not seeing fie=rst occurance
```

```python
mask= ~df["First Name"].duplicated(keep=False)
df[mask]
# this prints only unique values of First name in data frame
```

```python
df[mask].count()
#there are total values of 9 unique
```

```python
len(df[mask])
```

# The .drop_duplicates() Methods

This method drop duplicate values according to column wise.

```python
df=pd.read_csv("employees.csv",parse_dates=["Start Date","Last Login Time"])
df["Senior Management"]=df["Senior Management"].astype(bool)
df["Gender"]=df["Gender"].astype("category")
df.sort_values("First Name",inplace=True)
df.head(3)
```

```python
len(df)
```

```python
len(df.drop_duplicates())
```

```python
df.drop_duplicates(subset=["First Name"],keep="first")
# this makes first occured duplicates duplicates and droops all the other rows.
# if keep keep= first ,it keeps first name and if keep=last it keeps last occured value
```

```python
df.drop_duplicates(subset=["First Name"],keep=False)
# this drops the columns which are duplicate without considering first or last occured or how many times the row occured
```

```python
df.drop_duplicates(subset=["Team"],keep=False)
# no unique in team
```

```python
df.drop_duplicates(subset=["First Name","Team"])
#it's going to perform that duplicate dropping operation while looking at the values across two columns
# it don't care if the team is the same but the first name is different.
#Only if those two values are the same and those two columns it want to remove it. (same as and conditions)
```

```python
df.drop_duplicates(subset=["First Name","Team"],inplace=True)
df.head(22)
# here adam and araon are duplicate names but their team is unique so it prints them.
#Here the same follows to all the First Name elements.
```

```python
len(df) 
# by that reduction we get the memory reduced
```
