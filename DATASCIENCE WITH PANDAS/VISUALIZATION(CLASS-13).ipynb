# BAR GRAPHS

Bar graphs are done by using the methods according to the given conditions.

```python
goog=data.DataReader(name="GOOG", data_source="iex", start="2015-01-01", end="2018-11-26")
goog.head()
```

```python
def rank_performance(stock_price):
    if stock_price <=550:
        return "Poor"
    elif stock_price>550 and stock_price<=600:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
goog["close"].apply(rank_performance).value_counts()
```

```python
goog["close"].apply(rank_performance).value_counts().plot() #normal graph.
```

```python
goog["close"].apply(rank_performance).value_counts().plot(kind="Bar") #bar graph of close range
```

```python
def rank_performance_1(stock_price):
    if stock_price <=550:
        return "Poor"
    elif stock_price>550 and stock_price<=650:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
goog["high"].apply(rank_performance_1).value_counts().plot(kind="Bar") #bar graph of high range 
```

```python
goog["high"].apply(rank_performance_1).value_counts().plot(kind="Barh") # horizantal bar graph of high range 
```

# Pie chart


This is the representation of the data in some kinds of the graphs in the table.

```python
apple=data.DataReader(name="AAPL", data_source="iex",start="2015-01-01", end="2018-01-01")
apple.head()
```

```python
x=apple["close"].mean()
y=apple["open"].mean()
z=apple["low"].mean()
a=apple["high"].mean()
b=apple["volume"].mean()
```

```python
def rank_performance_3(stock_price):
    if stock_price>=z:
        return "Above Average"
    else:
         return "below Average"
```

```python
def rank_performance_4(stock_price):
    if stock_price>=y:
        return "Above Average"
    else:
         return "below Average"
```

```python
def rank_performance_5(stock_price):
    if stock_price>=a:
        return "Above Average"
    else:
         return "below Average"
```

```python
def rank_performance_6(stock_price):
    if stock_price>=b:
        return "Above Average"
    else:
         return "below Average"
```

```python
def rank_performance_2(stock_price):
    if stock_price>=x:
        return "Above Average"
    else:
         return "below Average"
```

```python
plt.style.use("dark_background")
apple["close"].apply(rank_performance_2).value_counts().plot(kind="pie",legend=True) # x defined.
```

```python
plt.style.use("dark_background")
apple["low"].apply(rank_performance_3).value_counts().plot(kind="pie",legend=True)  #z low defoned
```

```python
plt.style.use("dark_background")
apple["open"].apply(rank_performance_4).value_counts().plot(kind="pie",legend=True) #y open defined
```

```python
plt.style.use("dark_background")
apple["high"].apply(rank_performance_5).value_counts().plot(kind="pie",legend=True) #a high defined
```

```python
plt.style.use("dark_background")
apple["volume"].apply(rank_performance_6).value_counts().plot(kind="pie",legend=True) #b volume defined.
```

# HISTOGRAMS

This is same as pie and bar graph

```python
goog=data.DataReader(name="GOOG", data_source="iex",start="2015-01-01", end="2018-01-01")
goog.head()
```

```python
def custom_round(stock_price):
    return int(stock_price/100.0)*100
```

```python
custom_round(195.23) #rounds to smallest 100th's point
custom_round(95.23) #rounds to smallest 100th's point
custom_round(295.23) #rounds to smallest 100th's point
```

```python
goog["high"].apply(custom_round).value_counts().sort_index()
```

```python
goog["high"].apply(custom_round).nunique() 
#there are 9 unique high rounded values
```

```python
 goog["high"].apply(custom_round).plot(kind="hist") 
# this is the histogram but this donnot give actual measuaring due to duplicate values ,
#so we need to use bins="unique values" argumet to get perfect histogram.
#these 9 histograms represents the 9 set of high column  values
```

```python
goog["high"].apply(custom_round).plot(kind="hist", bins=9)  #using the bins argument to get perfectedgraph,recomnded always.
```
