# More String Methods - strip, lstrip, and rstrip
     This methods are likely to remove the white sopaces in string.
     l= left,r-right strip
     Any of this method does'nt remove the spaces in between the words

```python
chicago=pd.read_csv("chicago.csv").dropna()  # no null values.
chicago["Department"]=chicago["Department"].astype("category")  
chicago.tail(3)
```

```python
# Common method in strings.
"      Hello World".lstrip()  # this removes the withe space in left (or )begning
"Hello World     ".rstrip()  # this removes the withe space in right  (or )end
"     Hello World   ".strip() #this removes both left and begining spaces.
```

```python
# this method in pandas take input of every row of respected column and perform operation
chicago["Name"]=chicago["Name"].str.lstrip().str.rstrip() 
#together removes the both spaces like strip.
```

```python
chicago["Position Title"]=chicago["Position Title"].str.lstrip().str.rstrip()
# same as above query.
```

# The expand and n Parameters of the .str.split() Method

    In split method there will be expand parameter which helps us to make the "data frame" instead of "series of lists' if the expand=True,and by default the expand is set to false whic returns the "series".
    There is the n parameter to limit the number of columns of dataframe formed by the split() method.        

```python
chicago=pd.read_csv("chicago.csv").dropna()  # no null values.
chicago["Department"]=chicago["Department"].astype("category")  
chicago.tail(3)
```

```python
chicago["Name"].str.split(",", expand=True)
# here we can see that there is a "data frame" instead of  "series of lists"
```

```python
chicago[["First Name","Last Name"]]=chicago["Name"].str.split(",",expand=True)
# this code helps us to create new columns first and last name and has the stored value in the splited data frame.
# this is effect is done directly by the  original data frame.
chicago.head(3)
```

```python
chicago["Position Title"].str.split(" ",expand=True)
# this is mainly due to more number of spaces in the rows we can limit by using n parameter 
chicago["Position Title"].str.split(" ",expand=True,n=1)
#this limits the rows by zero - one column.
```
