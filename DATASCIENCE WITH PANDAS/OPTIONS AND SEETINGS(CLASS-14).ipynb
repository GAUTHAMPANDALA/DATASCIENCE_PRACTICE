# Introduction to the Options and Settings Module

```python
import pandas as pd
import numpy as np
```

# Changing pandas Options with Attributes and Dot Syntax

```python
data=np.random.randint(0,100,[1000,50])
#this returns the list of size 50 columns and 1000 rows
df=pd.DataFrame(data) 
#this makes the random number list to dataframes
df.head()
```

```python
 pd.options.display     #this shows all the options we can perform with the datafrmae
```

```python
pd.options.display.max_rows  #number of rows displayed by the cell(60)
```

```python
pd.options.display.max_rows=4 #this modifies above 60 to 4
```

```python
df
```

```python
pd.options.display.max_columns=10 #this modifies default columns to 4
```

```python
df
```

# Changing the Pandas Option with methods

Standard method to change the options of display.

Here we can change and set the options to new option.

```python
data=np.random.randint(0,100,[1000,50])
df=pd.DataFrame(data) 
df.head()
```

```python
pd.get_option("max_rows") #this gets the option of required method in above columns
```

```python
pd.get_option("max_columns")
```

```python
pd.set_option("max_rows",60) #sets rows from 4 to 60
```

```python
pd.set_option("max_column",50) #sets rows from 10 to 50
```

```python
pd.describe_option("max_columns") #describes the maxcolumns option.
pd.reset_option("max_columns")   #resets the maxcolumns option.
```

# The precision option


This option hides the floating number complexity and returns the simple number.

```python
df=pd.DataFrame(np.random.randn(5,5)) #this is random dataframe.
df
```

```python
pd.get_option("precision") #returns precision of each value in dataframe.
```

```python
pd.set_option("precision",2) 
#sets the precision of each value in dataframe from 6 to 2
```

```python
df
```

```python
pd.reset_option("precision") #resets the precesion back to original(6).
```

```python
df
```
