# Intro to the Module + Fetch Panel Dataset from IEX


Panels are the collection of the dataframes, this consists set of dataframes inside the panel.

Here we are going to see the three dimensional object called panels where as dataframe is two dimensional object and  the Series is the one dimensional object.

We need three point reference to access panel. 

1. Type of panel
2. TYPE OF dataframe inside panel
3. ROWS and Column details of the dataframe inside the panel.

The panel has:

1. item axis.
2. major axis.
3. minor axis.

We can  convert the dataframes into series and panels (vice-versa) and do required operations by converting them in 1,2,3 dimensional datastructures.

We can transpose swap the axes in dataframe and can store in new dataframe for further use.
NOTE:

IF incase we want to convert the dataframe to panel we use ".to_panel()" function, while converting the dataframe to panel we should has atleast 2 multi-indexes.

# Extracting with the .loc, .iloc, and .ix Methods (dataframe)

1. p.loc["index","column","columnvalueto extrct"]
2. p.iloc["indexnumber","column number"] -->counts the column number continusly
3. p.ix[""indexnumber/name","column number"]-->counts the column number continusly

```python
companies=["MSFT", "GOOG","AAPL","INTC","AMZN"]
p=data.DataReader(name=companies, data_source="iex", start="2015-01-01",end="2018-02-02")
p.head()
p.tail()
```

```python
p.loc["2015-01-08","open"]
#this locates 2015-01-08 open values
p.loc["2015-01-08","open"].AMZN
#this locates 2015-01-08's and inside we get "AMAZON open" value
```

```python
p.iloc[1,1]
#first parameter index second in numbering columns
# Iloc method counts each sub column with numbering 
#ie, open to volume has 0-24 numbering columns, total 25 columns.
p.iloc[1,24]
```

```python
p.ix["2015-01-08",5]
#first parameter index second in numbering columns
# Iloc method counts each sub column with numbering 
#ie, open to volume has 0-24 numbering columns, total 25 columns.
p.ix[777,0]
```

# Convert Panel to a MultiIndex DataFrame (and Vice Versa)


Here we convert dataframe to panel by multi-indexisng as prestep and panel to dataframe by ".to_frame()" method.

```python
companies=["MSFT", "GOOG","AAPL","INTC","AMZN"]
p=data.DataReader(name=companies, data_source="iex", start="2015-01-01",end="2018-02-02")
p.head()
p.tail()
```

```python
p=p.stack()# Here coverting   dataframe into multi index dataframe.
```

```python
p.head()
```

```python

# As  we have converted  dataframe to ulti index dataframe , now we can covert it into panel by this function.
p=p.to_panel()
```

```python
p #this isthe 3 dimensional object.
```

```python
p=p.to_frame() 
# now converted the panel back to dataframe. and so it has multi-index.
```

# The .swapaxes() Method (panels)

```python
companies=["MSFT", "GOOG","AAPL","INTC","AMZN"]
p=data.DataReader(name=companies, data_source="iex", start="2015-01-01",end="2018-02-02")
p.head()
p.tail()
p=p.stack()
p=p.to_panel()
p
```

```python
p.swapaxes("items","minor")  #this swaps the axis minor and items
```

```python
p2=p.swapaxes("items","minor")
```

```python
p.axes
```

```python
p2["AAPL"].head()
p2.minor_xs("close").head()
```

```python
p2.major_xs("2015-01-08")
```

# The .transpose() method

```python
companies=["MSFT", "GOOG","AAPL","INTC","AMZN"]
p=data.DataReader(name=companies, data_source="iex", start="2015-01-01",end="2018-02-02")
p.head()
p.tail()
p=p.stack()
p=p.to_panel()
p
```

```python
p.axes
```

```python
p.transpose(2,1,0)
# this makes items-->minor, major unchaged, minor-->items
```

```python
p.transpose(0,2,1)
# this makes items -->unchageds, major--> items, minor-->imajor 
```

```python
p.transpose(1,0,2)
# this makes items-->major,minor-->unchanged, major-->items
```
