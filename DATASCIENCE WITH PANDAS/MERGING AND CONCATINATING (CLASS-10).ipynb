# The pd.concat() Method, part-2

Here we are mainly going to see at the multi index process of the dataframe.

```python
week1=pd.read_csv("Restaurant - Week 1 sales.csv")
week2=pd.read_csv("Restaurant - Week 2 sales.csv")
customers=pd.read_csv("Restaurant - customers.csv")
foods=pd.read_csv("Restaurant - Foods.csv")
```

```python
sales=pd.concat([week1,week2], keys=["week 1","week 2"])
# here we get the multi index as week 1 and week 2 and corresponding indexes.
```

```python
sales.ix["week 1"]  # returns the multi index "WEEK 1"
sales.loc["week 2"]  # returns the multi index "WEEK 2"
```

```python
sales.ix[("week 1",2) ,"Food ID"] 
# returns outer layer week 1's inner layer 2's and their "foodid"
sales.ix[("week 2",4) ,"Customer ID"]
# returns outer layer week 1's inner layer 2's and their "customer id"
```

# INNER JOINS PART-1


The code we use is the "dataframe_name.merge()" function. It mainly has other-->"other dataframe to combine" ,how-->"which type of join is needed for that", on -->"on which conditional bias we need them to join", suffixes-->" to make the columns named properly"

This function works by equating the common customer ids in both tables and donot remove duplicate 

This is as same as the "sql join concept". 


```python
week1=pd.read_csv("Restaurant - Week 1 sales.csv")
week2=pd.read_csv("Restaurant - Week 2 sales.csv")
customers=pd.read_csv("Restaurant - customers.csv")
foods=pd.read_csv("Restaurant - Foods.csv")
```

```python
week1.merge(week2, how="inner", on="Customer ID" , suffixes=["-A","-B"],indicator=True,sort=True)
#merges the two dataframes by inner join on condition of dustomer id.
```

# THE OUTER JOINS


The outer join is the function that is used in "how" section of the merge function. This join is the combination of left,right join and both. We can also extact only left,right joined elements in the dataframe by this phenomenon.

Tjsi returns the values that exists in "either of tables or both of tables".
Here the "NAN" values represents the values that do not  present in the left (or) right tables.

This returns the dataframe of rows on which the condition is in both left and right columns, i.e the condition should be same in both rows ,and their values should not be same" and in left (or) right columns, but it do not select the rows that are equal in thier values.

There is other parameter called indicator which will be created in other column and teels us about the rows and their origin of rows.

This consits off both the tables values together only once. 

```python
week1=pd.read_csv("Restaurant - Week 1 sales.csv")
week2=pd.read_csv("Restaurant - Week 2 sales.csv")
customers=pd.read_csv("Restaurant - customers.csv")
foods=pd.read_csv("Restaurant - Foods.csv")
```

```python
week1.merge(week2, how="outer", on="Customer ID")
# the outer join is done by comparing both customer id of week 1 table to week 2 table.
```

```python
merged=week1.merge(week2, how="outer", on="Customer ID",indicator=True,sort=True) 
# this indicator denotes about the rows which are on the columns.
merged
```

```python
len(week1.merge(week2, how="outer", on="Customer ID",indicator=True))
# this means that the onlt 454 rows are present in either "week-1" or "week-2".
#There are 46 rows where we have a valid value for both the food ID in Week 1 and the food ID in week.
# ie,46 custom IDs that are found in both of them.
```

```python
merged["_merge"].value_counts()
# tis represents values only in left column are 195.
#                               right column are 197
# and there are 62 same conditional values (same as the "inner join part-1") which are  in both left,right columns.
#ie, there are 62 customer ids in common.
#TOTAL=454 ROWS
#and remaining 46 rows has the values not null and they are in both of tables.
```

```python
merged[merged["_merge"]=="left_only"] #returns only left rows.
len(merged[merged["_merge"]=="left_only"])
```

```python
merged[merged["_merge"]=="right_only"] #returns only right rows.
len(merged[merged["_merge"]=="right_only"])
```

```python
merged[merged["_merge"]=="both"] #returns onlyboth sides.
len(merged[merged["_merge"]=="both"])
```

```python
merged[merged["_merge"].isin(["left_only","right_only"])]
#returns the rows of left and right  tables
```

# Merging the indexes with left_index and right_index parameters

If we want to merge the tables in the bases of indexes (or) combination of tndexes of one table with columns of other, we use this parameters

Here we are merging in based on index positions (or) index columns

This works by searching the index_col  of first table to the same column of second table. i.e (index_colum of A=column of B)

This works only when we make the required column as indexes.

```python
week1=pd.read_csv("Restaurant - Week 1 sales.csv")
week2=pd.read_csv("Restaurant - Week 2 sales.csv")
customers=pd.read_csv("Restaurant - customers.csv",index_col="ID")
foods=pd.read_csv("Restaurant - Foods.csv",index_col="Food ID")
```

```python
foods.head()
```

```python
week1.head(3)
```

```python
customers.head(5)
```

```python
sales=week1.merge(customers, how="left", left_on="Customer ID",right_index=True)
#this merges by the  right index  week1 and  id column of the customers table right indxes.
```

```python
sales.head(3)
```

```python
sales1=customers.merge(foods, how="left", right_on="Food ID",left_index=True)
#this merges by the  left indexs week1 and id column of the sales.
```
