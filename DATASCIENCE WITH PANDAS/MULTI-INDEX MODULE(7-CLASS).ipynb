# Create a MultiIndex with the set_index() Method

  ie:set_index(key=["list of columns whiuch we weish to make them as index"])

```python
bigmac=pd.read_csv("bigmac.csv",parse_dates=["Date"]) 
bigmac.head(3)
```

```python
bigmac.set_index(keys="Date")
#sets the Date parameter as index.
#This is single string comprising of text which will be index for dataframe.
```

```python
bigmac.set_index(keys=["Date","Country"],inplace=True)
#making the multiindex in origina dataframe.
bigmac
```

# THE .stack() method
  
   Reshaping the data using stack() function in pandas converts the data into stacked format .i.e. the column is stacked row wise.i.e. the stack method which moved an index from the columns access to the rows access . When more than one column header is present we can stack the specific column header by specified the level. unstack() function in pandas converts the data into unstacked format.
   "here in this module carefully observe the outputs. here,"In stack(), there will be only one column but in that column it has more than one sub-row of values, which corresponds the original data frame values". Every time when this function executes  the columns will be as indexes and this again and again reshapes the data.
   
   "THE STACK METHOD MAKES ALL THE COLUMNS AS THE "INNER INDEX for the Existing index" FOR THE DATA FRAME".

```python
world=pd.read_csv("worldstats.csv",index_col=["country","year"])
world.head(3)
```

```python
world.stack() 
#Stack method makes the columns in world frame as inner indexes.
#ie. population and gdp column has became the inner indexes for Country and Year indexes.
#here, according to the rows perspective the size is doubles as we have reshaped the dataframe as output  
```

```python
world=world.stack().to_frame()  
#THE .toframe() function is used to get the reshaped data in proper reprsentation type.
world                          
#this overwrite's the original dataframe.
# Here the column is one but the Original columns such as Population and GDP are the sub rows of single column.
```

# THE .UNSTACK() METHOD-PART-I
  
   The .unstack method  moves an "multi-index layer of rows to the  columns axis".
   
   "CARE-FULLY OBSERVE THE OUTPUTS".
   
   This method makes the" Inner indexes to columns", i.e. this makes the multi-indexed column structure,At last we get back series but reversed series as the last and final outer index is unstacked.
   
   
   "The number.of inner columns=Number of unstacks"

```python
world=pd.read_csv("worldstats.csv",index_col=["country","year"])
world.head(3)
```

```python
s=world.stack().to_frame()
s.head(3)
```

```python
s=s.unstack().unstack().unstack() #This is 3 time unstack execution.


#1 st time , This "undo" the stack method, and makes Population,GDP back to column.
#and if we use other method it again makes the year index to the multiindexed column, 
#ie,in 2nd time, we get  population,Year are multi-indexed columns.
#  at 3rd time we get back series but reversed series as the last and final and outer index is also unstacked.
#Hint-->"(horizantal present separated->columns,vertical present line-->rows)""
      # "with 1st  unstack "Country-->year-->[columns: (Population, gDP)]""
      # "with 2nd  unstack "country-->[column: (Population,Gdp)-->Year] ""
      # "with 3rd  unstack  "(Population,gdp)-->Year-->country." -->" (series format with same indexing)""
      # "with 4th  unstack   "(Population,gdp)-->year-->[columns:Country]"
      # "With 5th unstack  "(Population,gdp)-->[columns:Country-->year]""
      # "with 6th   unstack   "[columns:Country-->year-->(population,Gdp)]"
      # "with 7th   unstack   "Country-->Year--> (Population,gdp)" "
# ""the above format is the original dataframe structure before "1st unstack",ie,by executing stack method in original frame"".
      # "with 8th  unstack  it gets back to 1st unstack executed output."
s
```

# THE .unstack() method part-II
   
   In the above code we have not given any arguments for moving row multi-index to column axis, so it start from the most inner index. 
     
  "The unstack method has the Level parameter and if we specify the index position,it migrates that respective row multi-index to column axis. The level may be in any of this  {-ve int,+ve int, even in row index name} ways just as arrays".
   
   
    "CAREFULLY OBSERVE THE OUTPUTS"

```python
world=pd.read_csv("worldstats.csv",index_col=["country","year"])
world.head(3)
```

```python
s=world.stack().to_frame()
s.head(3)
```

```python
s.unstack(2)  
# as specified the index 2, it migrated the (population,gdp) row multi index to column,we can specify ("2 or -1 or (population,gdp)")
s.unstack(1)  
# as specified the index 1, it migrated the (Year)) row multi index to column,we can specify ("1 or -2 or (Year)")
s.unstack(0)  
# as specified the index 1, it migrated the (Country)) row multi index to column we can specify ("0 or -3 or (Country)"")
```

# The .unstack() PART-III

   As, we know that we can pass some of parameters in the unstack method, we can even also fill the null values by integers, by using the "fill_val" parameter after the "level parameter separated by "," we can pass the list of index layer names", which we want to make them in column perspective, in that list also we can specify as any of this {-ve int,+ve int, even in row index name} ways.
  
   We have to do "n-1 names in list for n multi-indexes" while using .unstack() method by lists.

```python
world=pd.read_csv("worldstats.csv",index_col=["country","year"])
s=world.stack()
s.head(3)
```

```python
s.unstack(level=["country",1],fill_value=0)
#  AS, THERE ARE ONLY "3" "MULTI-INDEXES we have specified "2", i.e."Country,Year".
# "Fill_value"  fills the all NAN values in dataframe by" 0".
```

# The .pivot() method

   This pivot method is used for the orientation of data, This method is used to reduce the lenght of dataframe.This method makes the column values to column headers.
   
   ALWAYS IT IS OPTIMAL TO HAVE MORE UNIQUE COLUMN VALUES AS COLLUMN HEADERS.
  
   pivot(index="INDEX OF PIVOTED DATAFRAME(ALWAYS LESS UNIQUE COLUMN),COLUMNS="COLUMN HEADER FOR PIVOTED DATAFRAME",(MORE UNIQUE VALUES COLUMN", VALUES="VALUES FOR THAT COLUMN"(OR REMAINING COLUMNS OTHER THAN 'index,column parameter columns') ).

```python
sales=pd.read_csv("salesmen.csv", parse_dates=["Date"])
sales.head(3)
```

```python
len(sales)
```

```python
sales["Salesman"].value_counts()
# here we can see that there are  "5 unique columns in Salesman column" which are repeated in 1830 rows.
```

```python
sales=sales.pivot(index="Date", columns="Salesman", values="Revenue")
## So we make the Salesman column as the column header,and date as index, values as the revenue.
sales.head(3)
```

```python
len(sales)
# Here we can see that the no.of rows are reduced from "1830 to 366" by orienting the dataframe using" pivot" function.
```

# The .pivot_table() method
  
  This method takes the mormal datframe and returns the aggregated data frame.
  
  This method has some  parameters inside the function which is used to aggregate the data and return the data frame based on the "value" parameter, by "aggfunc" parameter ,and "index" parameter for indexes ,this may be a list, "columns" parameter for making the columns,this may be a list.
 
   We can aggregate the data in may ways like--> sum,mean,max,min, etc, and the dataframe can also be consits of single,multi indexes and single,multi columns. Here "Value" must be Integers and they may be of lists.
   
   "CAREFULLY OBSERVE THE OUTPUT".

```python
foods=pd.read_csv("foods.csv")
foods.head(3)
```

```python
foods.pivot_table(values="Spend",  index=["City","Gender"],  columns="Item", aggfunc="mean")
# as we can see that this returns the data frame with aggregated data of "spend" and aggredation by "mean", 
#and multi-indexes are "city,Gender" and columns with column headers  will be "Item's".
```

```python
foods.pivot_table(values="Spend",  index=["Gender","Item"],  columns="City", aggfunc="mean")
# as we can see that this returns the data frame with aggregated data of "spend" and aggredation by "mean", 
#and multi-indexes are  "Item's,Gender" and columns with column headers  will be"city".
```

```python
foods.pivot_table(values="Spend",  index=["City","Item"],  columns="Gender", aggfunc="mean")
# as we can see that this returns the data frame with aggregated data of "spend" and aggredation by "mean", 
#and multi-indexes are  "Item's,City" and columns with column headers  will be"Gender".
```

```python
foods.pivot_table(values="Spend",  index=["City","Item"],  columns="Frequency", aggfunc="mean")
# as we can see that this returns the data frame with aggregated data of "spend" and aggredation by "mean", 
#and multi-indexes are  "Item's,City" and columns with column headers  will be"Frequency".
```

# THE .pdmelt() method
  This method is  reverse of .piviot_table() method,takes aggregated table and gets the Normal dataframe. This function makes the  condeses the given  dataframe into the only 3 columns with index unchanged, the parameters inside the pd.melt are
  
  "Frame" represents the dataframe name, "id_vars" this represents the column which we are going to concentrate mainly on new data frame,"var_name" this is the integrated column which integrates all the column data into one format, "Value_name" is the value column which represents the "var_name column  values"  
  This mehod has only 3 columns 1.variable, ie all columns 2.value of corresponding columns in th variable section 3. main  column we specify in id_vars.

```python
sales=pd.read_csv("salesmen.csv", parse_dates=["Date"])
sales.head(3)
```

```python
pd.melt(sales, id_vars = "Salesman" ,var_name="Date and revenue", value_name="Revenue and Dates" ).head(35)
#Here  according to the id_var specified "salesman" has become the column 
#all other columns has become column elements in the "variable name"
# and the values of that column have becomed the elements in the "value column".
```
